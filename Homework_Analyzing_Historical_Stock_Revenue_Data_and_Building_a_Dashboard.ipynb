```python
!pip install yfinance

import yfinance as yf

# Extracting Tesla stock data
tesla_data = yf.download('TSLA', start='2022-01-01', end='2022-12-31')

# Reset index, save, and display the first five rows
tesla_data.reset_index(inplace=True)
tesla_data.to_csv('tesla_stock_data.csv', index=False)
print(tesla_data.head())
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the webpage containing the revenue data
url = 'https://finance.yahoo.com/quote/TSLA/financials?p=TSLA'

# Send a GET request to the webpage
response = requests.get(url)

# Parse the HTML content of the webpage
soup = BeautifulSoup(response.text, 'html.parser')

# Print the HTML content to inspect the structure
print(soup.prettify())

# Find the table containing the revenue data
revenue_table = soup.find('table', class_='W(100%) M(0)')

# Extract the revenue data from the table
revenue_data = []
if revenue_table:
    for row in revenue_table.find_all('tr'):
        data = [cell.text for cell in row.find_all('td')]
        revenue_data.append(data)
else:
    print("Revenue table not found.")

# Convert the revenue data into a DataFrame
if revenue_data:
    tesla_revenue = pd.DataFrame(revenue_data[1:], columns=revenue_data[0])
    # Display the last five rows of the dataframe
    print(tesla_revenue.tail())
else:
    print("No revenue data found.")
```
